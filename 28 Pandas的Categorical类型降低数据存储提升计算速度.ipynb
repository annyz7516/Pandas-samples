#### Pandas的Categorical类型降低数据存储提升计算速度

##### Pandas的Categorical类型利用“数字技巧”节省存储空间，提升运算速度

原始存储，字符串复制存储                                 Categorical存储，数字+查找表

         A                                                       0       数字字符串查找表
         B                                                       1          0: A
         C                                                       2          1: B
         B                                                       1          2: C
         A                                                       0     

#### 核心思路，在于将全部字符串存储变成数字中间存储，大幅降低存储空间

#### 1. 读取数据

```python
import pandas as pd
```

```python
df=pd.read_csv('./files/titanic.csv')
df.head()
```

```python
df.info()
```

```python
df.info(memory_usage='deep')
```

```python
df_cat=df.copy()
df_cat.head()
```

#### 2. 使用categorical类型降低存储量

```python
df_cat['Sex']=df_cat['Sex'].astype('category')
df_cat.info(memory_usage='deep')
```

```python
df_cat.head()
```

#### 3. 提升运算速度

```python
%timeit df.groupby('Sex').size()
```

```python
%timeit df_cat.groupby('Sex').size()
```
