```python
# 替换掉revenues和profits的前缀，数字分隔符和空值
df.loc[:, 'revenues'] = df['revenues'].str.replace('$','').str.replace(',','').str.replace('-','1').astype(float)
```

```python
df.loc[:, 'profits'] = df['profits'].str.replace('$','').str.replace(',','').str.replace('-','1').str.replace('(','').str.replace(')','').astype(float)
```

#### 3、相关系数和协方差
用途（超级厉害）：
1. 两只股票，是不是同涨同跌？程度多大？正相关还是负相关？
2. 产品销量的波动，跟哪些因素相关（正，负，程度？）

对于两个变量X，Y：
1. 协方差：衡量同向反向程度，如果协方差为正，说明X，Y同向，协方差越大说明同向程度越高；如果协方差为负，说明X，Y反向，协方差越小（绝对值越大）说明同向程度越高
2. 相关系数：衡量相似程度，当他们的相关系数为1时，说明两个变量变化的正向相似程度最大，当相关系数为-1时，说明反向相似度最大

```python
# covariance（协方差）矩阵：
df.cov()
```

```python
# correlation coefficient (相关系数：标准化后的协方差，值在-1到1之间，更重要！！！)矩阵：
df.corr()
#
```

```python
# 单独查看revenue和profit的相关系数
df['revenues'].corr(df['profits'])
```

```python
# 单独查看revenue和expense(revenue-profit)的相关系数
df['revenues'].corr(df['revenues']-df['profits'])
```
