#### Pandas数据排序
Series的排序：
##### Series.sort_values(ascending=True, inplace=False)
参数说明：
+ ascending：默认True升序，False降序
+ inplace：是否修改原始Series

DataFrame的排序：
##### DataFrame.sort_values(by,ascending=True,inplace=False)
参数说明：
+ by：字符串或者list<字符串>，单列排序或者多列排序
+ ascending：bool或者List，升序还是降序，如果是list对应by的多列
+ inplace：是否修改原始DataFrame


```python
import pandas as pd
```

#### 读取数据

```python
fpath = "C:/Users/pxpxz_ct9p1p3/Downloads/Fortune_1000_Companies_by_Revenue.csv"
df = pd.read_csv(fpath)
```

```python
# 替换掉前缀$等
df.loc[:, 'revenues'] = df['revenues'].str.replace('$', '').str.replace(',','').astype('float')
# 替换掉前缀$等
df.loc[:, 'profits'] = df['profits'].str.replace('$', '').str.replace(',','').str.replace('(','').str.replace(')','').str.replace('(','').str.replace('-','0').astype('float')
```

#### 2、DataFrame的排序
##### 2.1 单列排序

```python
df.sort_values(by='revenues')
```

```python
df.sort_values(by='revenues', ascending=False)
```

#### 2.2 多列排序

```python
# 按profit，revenue排序，默认升序
df.sort_values(by=['profits', 'revenues'])
```

```python
# 两个字段都是降序
df.sort_values(by=['profits', 'revenues'],ascending=False)
```

```python
# 两个字段一升一降
df.sort_values(by=['profits', 'revenues'],ascending=[True,False])
```
