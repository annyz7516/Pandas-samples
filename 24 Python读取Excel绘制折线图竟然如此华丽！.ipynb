#### Python怎样结合Pyecharts绘制交互性折线图？
背景：
+ Pandas是Python用于数据分析领域的超级牛的库
+ Echart是百度开源的非常好用强大的可视化图表库，Pyechart是它的Python库版本

#### 1. 读取数据

```python
import pandas as pd
```

```python
df=pd.read_csv('./files/bidu_stock.csv',index_col='日期',parse_dates=True)
df.head()
```

```python
df.index
```

```python
df.sort_index(inplace=True) # inplace=True 代表修改这个df！！！
df.head()
```

#### 2.使用Pyecharts绘制折线图

```python
# 如果没有安装，使用pip install pyecharts安装
from pyecharts.charts import Line
from pyecharts import options as opts
```

```python
# 折线图
line=Line()

# x轴 to_list() 变成列表
line.add_xaxis(df.index.to_list())

# 每个y轴
line.add_yaxis('开盘价',df['开盘'].round(2).to_list())
line.add_yaxis('收盘价',df['收盘'].round(2).to_list())

# 图标配置
line.set_global_opts(
    title_opts=opts.TitleOpts('bidu_stock_2023'),
    tooltip_opts=opts.TooltipOpts(trigger='axis',
    axis_pointer_type='cross')
)
```

```python
# 渲染数据
line.render_notebook()
```
